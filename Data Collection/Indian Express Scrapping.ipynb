```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
page = 1
titles = []
times = []
while page != 182:
      url = f"https://indianexpress.com/section/business/banking-and-finance/page/{page}/"
      response = requests.get(url)
      html = response.content
      soup = BeautifulSoup(html, "html.parser")
      for h2 in soup.find_all("h2", class_="title"):
            titles.append(h2.get_text(strip=True))
      
      for time in soup.find_all('div', class_='date'):
            times.append(time.get_text(strip = True))
        
      page = page + 1
```

```python
import pandas as pd
IndianExpress = pd.DataFrame([titles, times])
IndianExpress = IndianExpress.T
```
