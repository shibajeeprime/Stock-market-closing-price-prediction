```python
import time
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.keys import Keys


browser = webdriver.Chrome('/home/sysadm/Downloads/chromedriver_linux64/chromedriver')
browser.get("https://economictimes.indiatimes.com/industry/banking/finance/banking")
time.sleep(1)

elem = browser.find_element_by_tag_name("body")

t = 0
while(t<180):
    try:
        loadmorebutton = browser.find_element_by_class_name("autoload_continue")
        time.sleep(1)
        loadmorebutton.click()
        time.sleep(1)
        t+=1
    except Exception as e:
        print(e, 'E')
        print(type(loadmorebutton))
        print(t)
        if t >= 155:
            break
        else:
            continue
     
    
```

```python
browser.execute_cdp_cmd('Emulation.setScriptExecutionDisabled', {'value': True})
post_elems = browser.find_elements_by_class_name("eachStory")

headline = []
for post in post_elems:
    head = post.find_elements_by_tag_name("h3")
    for h in head:
        headline.append(h.text)
timeline = []
for post in post_elems:
    time = post.find_elements_by_class_name("time")
    for t in time:
        timeline.append(t.text)

EconomicTimes = pd.DataFrame([headline, timeline])
EconomicTimes = EconomicTimes.T
EconomicTimes
```
