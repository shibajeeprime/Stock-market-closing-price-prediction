```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
page = 1
titles = []
times = []
while page != 83:
      url = f"https://www.business-standard.com/topic/banking-sector/{page}"  
      response = requests.get(url)
      html = response.content
      soup = BeautifulSoup(html, "html.parser")
      for li in soup.find_all("div", class_="listing-txt"):
            for h2 in li.find_all('h2'):
                titles.append(h2.get_text(strip=True))
            for time in li.find('p', class_ = ''):
                times.append(time)
        
      page = page + 1
```

```python
BusinessStandard = pd.DataFrame([titles, times])
BusinessStandard  = BusinessStandard .T
```
