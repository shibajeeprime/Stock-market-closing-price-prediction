```python
import time
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.keys import Keys


browser = webdriver.Chrome('/home/sysadm/Downloads/chromedriver_linux64/chromedriver')
browser.get("https://www.businesstoday.in/industry/banks")
time.sleep(1)

elem = browser.find_element_by_tag_name("body")

t = 0
while(t<60):
    try:
        loadmorebutton = browser.find_element_by_id("load_more")
        time.sleep(1)
        loadmorebutton.click()
        time.sleep(1)
        t+=1
    except Exception as e:
        print(e, 'E')
        print(type(loadmorebutton))
        print(t)
        if t >= 48:
            break
        else:
            continue
     
```

```python
browser.execute_cdp_cmd('Emulation.setScriptExecutionDisabled', {'value': True})
post_elems = browser.find_elements_by_class_name("widget-listing")

headline = []
for post in post_elems:
    head = post.find_elements_by_tag_name("h2")
    for h in head:
        headline.append(h.text)
timeline = []
for i in range(0, 21):
    times = post_elems[i].find_elements_by_tag_name("span")
    for t in times:
        timeline.append(t.text)
for i in range(21, len(post_elems)):
    times = post_elems[i].find_elements_by_tag_name("h5")
    for t in times:
        timeline.append(t.text)


BusinessToday = pd.DataFrame([headline, timeline])
BusinessToday = BusinessToday.T
BusinessToday
```

```python
BusinessToday.to_csv('/home/sysadm/Documents/BusinessToday.csv', header = False)
```
