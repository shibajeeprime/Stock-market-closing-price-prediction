```python
S=['Business_Today','Business_Standard','Economic_Times','Indian_Express']
```

```python
for s in S:
    d1=pd.read_csv('sentiment/'+s+'.csv')
    d1=d1.drop(['Sentiment'],axis=1)
    H=d1['Headlines']
    H2=[]
    for h in H:
        org_string = h
        new_string = re.sub(r'[^\w\s]','',org_string)
        new_string = re.sub(r"[a-zA-Z]",
                        lambda x :  x.group(0).lower(),
                        new_string)
        H2.append(new_string)
    d1['Headlines']=H2
    d1.to_csv('sentiment/'+s+'.csv',index=False)
```

```python
for t in S:
    d1=pd.read_csv('sentiment/'+t+'.csv')
    H=d1['Headlines']
    H3=[]
    for h in H:
        text=h
        for s in v:
            text = re.sub(r"\b{}\b".format(s), "", text)
            result = re.sub(r"\s+"," ", text, flags = re.I)
        H3.append(result)
    d1['Headlines']=H3
    d1.to_csv('sentiment/'+t+'.csv',index=False)
```
