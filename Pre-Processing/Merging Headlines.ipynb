```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/home/sysadm/Downloads/MoneyControl  - MoneyControl .csv', header = None); data
```

```python
data = data.iloc[1:,:]; data
```

```python
data.columns = ['i','Index', 'Headlines', 'Date']; data
```

```python
data =data.drop(['i','Index'], axis = 1); data
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y'); data
```

```python
S = []
for i in data['Date'].unique():
    l1 = data["Date"] == i
    l2 = data["Headlines"][l1]
    s =""
    for i in l2:
        s1 = i
        s = ". ".join([s1,s])
    S.append(s)
        
```

```python
d = data['Date'].unique(); d
```

```python
MoCo = pd.DataFrame([S,d])
MoCo = MoCo.T
MoCo.columns = ['Headlines', 'Date']; MoCo
```
