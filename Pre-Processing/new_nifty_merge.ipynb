```python
import numpy as np
import pandas as pd
```

```python
eco = pd.read_csv('Indian_Express.csv')
eco.drop(columns=eco.columns[0], axis=1, inplace=True)
eco['Date']=pd.to_datetime(eco['Date'], format='%Y-%m-%d'); eco
```

```python
sent=[]
for j in range(len(eco)):
  sent.append([float(i) for i in eco["Sentiment"][j][2:-2].split(" ") if i!=""])
```

```python
len(sent)
```

```python
sent
```

```python
F = pd.DataFrame(eco)
F['Sentiment'] = sent;F
```

```python
market_data = pd.read_csv('NIFTY BANK_Data.csv'); market_data 
```

```python
market_data['Date']=pd.to_datetime(market_data['Date'], format='%d-%b-%y'); market_data
```

```python
merged_df = market_data.merge(F, how = 'left', on = ['Date']); merged_df
```

```python
merged_df = merged_df.fillna(''); merged_df
```
