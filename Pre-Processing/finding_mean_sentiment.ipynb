```python
import numpy as np
import pandas as pd
```

```python
eco = pd.read_csv('/content/Economic_Times.csv')
eco.drop(columns=eco.columns[0], axis=1, inplace=True)
eco['Date']=pd.to_datetime(eco['Date'], format='%Y-%m-%d'); eco
```

```python
sent=[]
for j in range(len(eco)):
  sent.append([float(i) for i in eco["Sentiment"][j][2:-2].split(" ") if i!=""])
```

```python
d1=pd.read_csv('/content/Bus_std_Nifty_merged.csv')
d2=pd.read_csv('/content/Bus_tdy_Nifty_merged.csv')
d3=pd.read_csv('/content/Ind_Exp_Nifty_merged.csv')
d4=pd.read_csv('/content/eco_tm_Nifty_merged.csv')
```

```python
d1
```

```python
d1=d1.drop(d1.columns[0],axis=1)
d2=d2.drop(d2.columns[0],axis=1)
d3=d3.drop(d3.columns[0],axis=1)
d4=d4.drop(d4.columns[0],axis=1)
```

```python
S1=d1['Sentiment']
S2=d2['Sentiment']
S3=d3['Sentiment']
S4=d4['Sentiment']
```

```python
for  j in range(len(S1)):
  if S1.isna()[j]==False:
    s=S1[j].replace('[','').replace(',','').replace(']','').split()
    for i in range(len(s)):
      s[i]=float(s[i])
    S1[j]=s

for  j in range(len(S2)):
  if S2.isna()[j]==False:
    s=S2[j].replace('[','').replace(',','').replace(']','').split()
    for i in range(len(s)):
      s[i]=float(s[i])
    S2[j]=s

for  j in range(len(S3)):
  if S3.isna()[j]==False:
    s=S3[j].replace('[','').replace(',','').replace(']','').split()
    for i in range(len(s)):
      s[i]=float(s[i])
    S3[j]=s

for  j in range(len(S4)):
  if S4.isna()[j]==False:
    s=S4[j].replace('[','').replace(',','').replace(']','').split()
    for i in range(len(s)):
      s[i]=float(s[i])
    S4[j]=s

    
```

```python
for j in range(len(S1)):
  if S1.isna()[j]==False:
    s=np.array(S1[j])
    S1[j]=s

for j in range(len(S2)):
  if S2.isna()[j]==False:
    s=np.array(S2[j])
    S2[j]=s

for j in range(len(S3)):
  if S3.isna()[j]==False:
    s=np.array(S3[j])
    S3[j]=s

for j in range(len(S4)):
  if S4.isna()[j]==False:
    s=np.array(S4[j])
    S4[j]=s
```

```python
for j in range(len(S1)):
  if S1.isna()[j]==True:
    S1[j]=np.NaN

for j in range(len(S2)):
  if S2.isna()[j]==True:
    S2[j]=np.NaN

for j in range(len(S3)):
  if S3.isna()[j]==True:
    S3[j]=np.NaN

for j in range(len(S4)):
  if S4.isna()[j]==True:
    S4[j]=np.NaN
```

```python
df=S1[0:7]
df=[np.mean(df) if i is np.NaN else i for i in df]
S1[0:7]=df
df=S2[0:7]
df=[np.mean(df) if i is np.NaN else i for i in df]
S2[0:7]=df
df=S3[0:7]
df=[np.mean(df) if i is np.NaN else i for i in df]
S3[0:7]=df
df=S4[0:7]
df=[np.mean(df) if i is np.NaN else i for i in df]
S4[0:7]=df
```

```python
for i in range(7,len(S1)):
  if S1.isna()[i]==True:
    S1[i]=np.mean(S1[i-7:i-1])

for i in range(7,len(S2)):
  if S2.isna()[i]==True:
    S2[i]=np.mean(S2[i-7:i-1])

for i in range(7,len(S3)):
  if S3.isna()[i]==True:
    S3[i]=np.mean(S3[i-7:i-1])

for i in range(7,len(S4)):
  if S4.isna()[i]==True:
    S4[i]=np.mean(S4[i-7:i-1])
```

```python
zipped=list(zip(S1,S2,S3,S4))
SD=pd.DataFrame(zipped, columns=['Bus_std','Bus_tdy','Ind_exp','eco_tm'])
SD
```
